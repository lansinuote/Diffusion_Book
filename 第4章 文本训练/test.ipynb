```python
#第4章/定义测试函数
from diffusers import StableDiffusionPipeline
from matplotlib import pyplot as plt
%matplotlib inline
import torch


def test(pipeline):
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    pipeline = pipeline.to(device)

    texts = [
        'a photo of a <cat-toy>', 'a photo of a clean <cat-toy>',
        'a photo of a dirty <cat-toy>', 'a dark photo of the <cat-toy>'
    ]
    images = []
    for text in texts:
        image = pipeline(text, num_inference_steps=25).images[0]
        images.append(image)

    plt.figure(figsize=(20, 10))
    for i in range(4):
        plt.subplot(1, 4, i + 1)
        plt.imshow(images[i])
        plt.axis('off')

    plt.show()
```

```python
#第4章/测试训练前的模型
pipeline = StableDiffusionPipeline.from_pretrained(
    'runwayml/stable-diffusion-v1-5', safety_checker=None)

test(pipeline)
```

```python
#第4章/测试训练后的模型
pipeline = StableDiffusionPipeline.from_pretrained('./save',
                                                   safety_checker=None)

test(pipeline)
```

```python
#第4章/在线加载笔者训练好的模型并测试
pipeline = StableDiffusionPipeline.from_pretrained(
    'lansinuote/diffusion.2.textual_inversion', safety_checker=None)

test(pipeline)
```
